#### Get the libs for this project:

```python
import numpy as np
import pandas as pd
import geopandas as gpd

import matplotlib.pyplot as plt
from shapely.geometry import box, Polygon, LineString, Point

import os

data_folder = "Resources/01_Data"
```

#### Load Files

```python
# pandas data frame from cites in a csv
df = pd.read_csv("data_files/ufo_data.csv")

print(df.head())

#gdf.set_crs(epsg=4326, inplace=True)
```

We need to convert the lat/lon columns to a `geometry` data type. A `Point` to be more specific. Using our csv file and the dataframe (`df1`), we create a `geodataframe` (`gdf1`) using the coordinates in `df1`

```python
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat))
gdf.set_crs(epsg=4326, inplace=True)
gdf.head()
```

```python

#boundary = gpd.read_file("data/us_nation_border.geojson")

us_borders_shp = os.path.join(data_folder,'country_and_city_data/us_border_shp/us_border.shp')
us_borders_geo = os.path.join(data_folder,'country_and_city_data/us_borders.geojson')

boundary = gpd.read_file(us_borders_shp)
borders = gpd.read_file(us_borders_geo)

fig, ax = plt.subplots(figsize=(12, 10))

# for i, row in borders.iterrows():
#     #print(row['geometry'])
#     row.plot(ax=ax,color="gray")

# 
#boundary.plot(ax=ax, color="gray")
borders.plot(ax=ax,color="gray")
gdf.plot(ax=ax, markersize=0.1, color="blue")
ax.axis("off")
plt.axis('equal')

minx, miny, maxx, maxy = boundary.total_bounds
print(minx,miny,maxx,maxy)
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
plt.show()
```
