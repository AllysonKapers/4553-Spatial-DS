## Osmnx Is Not My Friend

```python
import osmnx as ox


orig = (-97.4487, 35.2636)
dest = (-95.9985, 36.1734)

#osmnx.graph.graph_from_bbox(north, south, east, west, network_type='all_private', simplify=True, retain_all=False, truncate_by_edge=False, clean_periphery=True, custom_filter=None)

#G = ox.graph_from_point(orig, distance=241402, simplify=True, network_type="drive") 
G = ox.graph_from_bbox(36.1734, 35.2636, -95.9985, -97.4487, network_type='drive', simplify=False,  custom_filter='["highway"~"motorway"]')
```

```python
import networkx as nx
#print(dir(G))

# print(dir(ox))

origin_node = ox.get_nearest_node(G, orig) 
destination_node = ox.get_nearest_node(G, dest) 
route = nx.shortest_path(G, origin_node,destination_node) 
print(route)
```

```python
#fig, ax = ox.plot_graph_route(G, route, route_linewidth=6, node_size=0, bgcolor='k')
fig, ax = ox.plot_graph_route(G, route)
```

```python
fig, ax = ox.plot_graph(G, show=False, close=False)
```
